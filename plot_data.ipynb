```python
import pandas as pd
from pathlib import Path
```

```python
current_dir = Path.cwd()
dataset_dir = current_dir / 'Original Dataset'

df_metadata = pd.read_excel(dataset_dir / 'Test-Variablen Software ab 2014.xlsx')
df_pre_therapy_ratings = pd.read_excel(dataset_dir / 'Reha BG 2014-2022 prä Daten+Ther.ratings_anonym(1).xlsx')
df_post_therapy_ratings = pd.read_excel(dataset_dir / 'Reha BG 2014-2022 post Daten_anonym(1).xlsx')
```

```python
if df_metadata['Variablenname'].duplicated().any():
    metadata_df = df_metadata.drop_duplicates(subset=['Variablenname'], keep='first')
    
if df_pre_therapy_ratings.columns.duplicated().any():
    df_pre_therapy_ratings = df_pre_therapy_ratings.loc[:, ~df_pre_therapy_ratings.columns.duplicated(keep='first')]

if df_post_therapy_ratings.columns.duplicated().any():
    df_post_therapy_ratings = df_post_therapy_ratings.loc[:, ~df_post_therapy_ratings.columns.duplicated(keep='first')]
```

```python
# Lookupup-Table: Mapping from Codes to desired metadata 
metadata_lookup = metadata_df.set_index('Variablenname')['Variablenlabel']
sorted_metadata = metadata_lookup.reindex(df_pre_therapy_ratings.columns)
    
# Create MultiIndex from metadata
multi_index = pd.MultiIndex.from_arrays(
    [
        sorted_metadata.fillna('Unbekannt: Variablenlabel'),  # Layer 1: Metadata labels
        sorted_metadata.index # Layer 2: Original variable codes
    ],
    names=['Variablenlabel', 'Code']
)

df_pre_therapy_ratings.columns = multi_index
print(df_pre_therapy_ratings.head())
```
